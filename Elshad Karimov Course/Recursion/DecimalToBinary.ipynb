```python
def DecimalToBinary(number):
  '''
  To convert a number for decimal to binary we need to get the remainder of a division plus ten and multiply recursively over the results divided by two. 

  In python the remainder can be taken using the % (percentage) operator.
  
  '''
# unintentional case
  assert int(number) == number, 'The number must be integer' 
# base case or the stop criterion or infinity loop prevention
  if number == 0:
    return 0
# recursive case
  return number % 2 + 10 * DecimalToBinary(int(number / 2))
```
