```python
def fibonacci(number):
  '''
    # Fibonacci numbers

    ## Recursive case
    ### Fibonacci number is a sequence of numbers that sums up from 2 preceding numbers starting from 0 and 1
    ### 0, 1, 1, 2, 3, 5, 8, 13...
    ### Fibonacci of 4 is equal to 3
    ### Fibonacci of 5 is equal to 5
    ### Fibonacci of 6 is equal to 8
  
  '''
  # the unintentional cases
  assert number >= 0 and int(number) == number, 'The number must be positive integer'
  # the stop criterion or the base case
  if number in [0,1]:
    return number
  # the recursive case 
  else:
      return fibonacci(number - 1) + fibonacci(number - 2)
```

```python
fibonacci(6)
```
