```python
# In case you need to increase the recursion limit
# if you need to do it maybe the stop criterion is wrong or there is a huge calculation to do
import sys
sys.setrecursionlimit(44)
```

```python
# Factorial problem
def factorial(number):
  """
  
  ## First step > our recursive case: n! = n * (n-1) * (n-2) * ...
  ## Second step > the stop criterion: 0! = 1 and 1! = 1, then n in [0,1] = 1
  
  """
  assert number >= 0 and int(number) == number, 'The number must be positive integer only'
  if number in [0,1]:
    return 1
  else:
    return number * factorial(number-1)  
```
