```python
def PowerOfNumber(Number, Exponential):
  
    '''
    In v2 an unintentional case could be negative or fractional negative or decimal negative number
    So, developing a way for solving it using simple math
         if Exponential < 0:
         return (1 / Number) * PowerOfNumber(Number, (Exponential*-1) - 1)

    '''
    # unintentional cases
    assert Exponential >= 0 and int(Exponential) == Exponential, 'The Exponential must be a positive integer'
    # base case or the stop criterion
    if Exponential == 0:
      return 1
    if Exponential == 1:
      return Number
    # recursive case
    else:
      return Number * PowerOfNumber(Number, Exponential - 1)
```
