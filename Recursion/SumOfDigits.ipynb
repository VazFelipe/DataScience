```python
def SumOfPositiveNumber(PositiveNumber):
  '''
    It sums up the digits of a particular positive integer number.

    It separates the digits recursively using the floor division and the remainder 

    54 // 10 = 5 with remainder 4
    54 % 10  = 4 (the remainder)

    (54 // 10) + (54 % 10) = 9 

    9   = 9
    10  = 1 + 0 = 1
    11  = 1 + 1 = 2
    100 = 1 + 0 + 0 = 1 
    101 = 1 + 0 + 1 = 2

  '''
  
  # constraint, prevents unintentional cases, here negative and fractional numbers 
  assert PositiveNumber >= 0 and int(PositiveNumber) == PositiveNumber, 'The number must be a positive integer'
  # base case, prevents it goes to infinity
  if PositiveNumber == 0:
    return 0
  # recursive case
  else:
    return int(PositiveNumber%10) + SumOfPositiveNumber(int(PositiveNumber//10))
```
