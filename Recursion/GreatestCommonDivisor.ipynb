```python
def gcd(number1, number2):
  '''
  GCD is the greatest divisor between two positive integer numbers. 
  
  For example:
    gcd(8, 12) = 4 --> is we factorate 8 and 12 we will see that:
  
         8 = 2 x 2 x 2
        12 = 2 x 2 x 3
  
    So, both numbers has the first two multipliers as 2 and 2 times 2 is 4.
    This is how the GCD works.

  To implement gcd(number1, number2) we could use the Euclidean algorithm. 
  
  For example:
    gcd(48, 18) = 48 / 18 = 2 remainder 12
                  18 / 12 = 1 remainder 6
                  12 /  6 = 2 remainder 0
  
    In this case, 6 is the Greatest Common Divisor for 48 and 12!  

  gcd(number1, number2) works as Euclidean Algorithm onlyfor posivite integer numbers.

  In Python the remainder can be taken using % (percentage) operator.

  '''
  # unintentional cases
  assert number1 >= 0 and number2 >= 0 and int(number1) == number1 and int(number2) == number2, 'The number1 and number2 must be a positive integer number.'
  # base case
  ## here is the Euclidean Algorithm, see more details in the helper "? gcd()". Run without "". 
  if number2 == 0:
    return number1
  # recursive case
  ## here is recursive case getting the remainder and dividing by number2
  else:
    return gcd(number2, number1%number2)
```
